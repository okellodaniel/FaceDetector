```python
# Basics of working with OpenCV

'''
Importation of Libraries
'''
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

'''
Reading an image
'''
img_raw = cv2.imread('./mandrill_colour.png')

# Type and Shape of the Array

type(img_raw)
np.ndarray

img_raw.shape
(1300,1950,3)

# Viewing the Image

plt.imshow(img_raw)

'''
- OpenCv shows img in RGB
- MatplotLib displays image in BGR

=> That is why there is the Bluish color

'''

```
