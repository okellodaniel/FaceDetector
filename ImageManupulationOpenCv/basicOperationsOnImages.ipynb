```python
'''
Import Libraries
'''
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2


'''
Creating a black Image which will act as a template
'''

image_blank = np.zeros(shape=(512,512,3), dtype=np.int16)

'''
Display black Image
'''

plt.imshow(image_blank)
```

```python
'''The generalised function for drawing shapes on Images is'''

# cv2.shape(line,rectangle etc)(image, Pt1, Pt2, color, thickness)

'''
=> Argumanets here
---------------------------------------
- Image on which shapes are to be drawn
- co-ordinates of the shape to be drawn from Pt1(top left) to Pt2(bottom right)
- Color: The color of the shape that is to be drawn. It is passed as a tuple, 
   eg: (255,0,0). For grayscale, it will be the scale of brightness.
- The thickness of the geometrical figure.
'''
```

```python
'''Straight Line Across  an Image'''

# Draw a diagonal line with a thickness of 5px

line_red = cv2.line(image_blank,(0,0),(511,511), (255,0,0), 5)
plt.imshow(line_red)
```

```python
'''Drawing a diagonal green line with thickness 5px'''

green_line = cv2.line(image_blank, (0,0),(511,511),(0,255,0),5)
plt.imshow(green_line)
```

```python
'''Writing on Images'''

'''
* Adding Text needs specifying different Argments
    - Text to be written
    - Coordinates of the text. The text on an image
      begins from the bottom left direction.
    - Font type and Scale
    - Other Attributes like Color, Thickness, Line-Type
    => Line type used is => lineType = cv2.LINE_AA
'''

font = cv2.FONT_HERSHEY_SIMPLEX

#                   image      Text   Coordinates type scale color   other attributes   
text = cv2.putText(image_blank,'OpenCV',(10,500),font,4,(255,255,255),2,cv2.LINE_AA )
plt.imshow(text)
```
