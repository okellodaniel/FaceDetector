```python
#Import out Libraries

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

# Loading the saved Image

test_img = cv2.imread('./img.jpeg')

# Coverting to grey scale

test_img_gray = cv2.cvtColor(test_img, cv2.COLOR_BGR2GRAY)

# Displaying grayscale Image

plt.imshow(test_img_gray , cmap='gray')
```

```python
'''OpenCv loads Image in BGR format, covert this
 to RGB format'''

def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
'''Loading the Classifier for Frontal face'''

haarCascadeFace = cv2.CascadeClassifier('data/haarcascade_frontalface_default.xml')
```

```python
'''Detect Multiscale module of Classifier is used

        # Parameters to be used here will be
    ==============================================
    1. ScaleFactor
    2. MinNeighbors
'''
```

```python
faces_rects = haarCascadeFace.detectMultiScale(test_img,scaleFactor=1.2,minNeighbors=5)

#Print Number of Faces found

print('Faces Found :', len(faces_rects))
```

```python
'''
Loop over all Coordinates returned and draw rectangles around
 them using OpenCV
'''

for(x,y,w,h) in faces_rects:
    cv2.rectangle(test_img,(x,y),(x+w,y+h),(0,255,0),2)
```

```python
'''
Display original Image in Colored to show detected face
'''

plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
```

```python
'''Face-Detection with Generalised Function
===============================================
'''

def detect_faces(cascade,test_img,scalefactor=1.1):
    # Create Image copy to prevent any changes to Original One
    img_copy = test_img.copy()

    '''Convert test Image to grayscale as OpenCv facedetect 
        expects gray images
    '''
    gray_img = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)

    # Appy haar classifier

    faces_rect = cascade.detectMultiScale(gray_img,scaleFactor=scalefactor,minNeighbors=5)

    for(x,y,w,h) in faces_rect:
        cv2.rectangle(img_copy,(x,y),(x+w,y+h),(0,255,0),2)
    
    return img_copy
```

```python
'''
Use Generalised Function to detect faces
'''

test_img2 = cv2.imread('./test2.jpeg')

# Converting to GrayScale

test_img_gray = cv2.cvtColor(test_img2, cv2.COLOR_BGR2GRAY)

# Display GrayScale

plt.imshow(test_img_gray, cmap="gray")

```

```python
# Call Function to detect faces

faces = detect_faces(haarCascadeFace,test_img2)

# Covert RGB and Display Image

plt.imshow(cv2.cvtColor(faces, cv2.COLOR_BGR2RGB))
```
